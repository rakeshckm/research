## Data preprocessing for vision transformer

### Subtask:
Prepare the image data and labels for input into a vision transformer model. This may involve loading images, resizing, normalizing, and encoding labels.


**Reasoning**:
Import the necessary libraries for image preprocessing and label encoding.



```python
import os
from PIL import Image
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
```

**Reasoning**:
Define a function to load and preprocess images, and then apply this function to the image paths in the training, test, and validation dataframes.



```python
def load_and_preprocess_image(image_path, target_size):
    try:
        img = Image.open(image_path).convert('RGB')
        img = img.resize(target_size)
        img_array = np.array(img)
        img_array = img_array / 255.0
        return img_array
    except Exception as e:
        print(f"Error processing image {image_path}: {e}")
        return None

target_size = (128, 128)

train_images = [load_and_preprocess_image(path, target_size) for path in train_df['image_path']]
test_images = [load_and_preprocess_image(path, target_size) for path in test_df['image_path']]
val_images = [load_and_preprocess_image(path, target_size) for path in val_df['image_path']]

# Remove None values in case of errors during image loading
train_images = [img for img in train_images if img is not None]
test_images = [img for img in test_images if img is not None]
val_images = [img for img in val_images if img is not None]
```

**Reasoning**:
Encode the string labels into numerical labels using LabelEncoder and convert the preprocessed image lists into NumPy arrays.



```python
label_encoder = LabelEncoder()

train_labels_encoded = label_encoder.fit_transform(train_df['label'])
test_labels_encoded = label_encoder.transform(test_df['label'])
val_labels_encoded = label_encoder.transform(val_df['label'])

train_images_array = np.array(train_images)
test_images_array = np.array(test_images)
val_images_array = np.array(val_images)

print("Shape of preprocessed training images:", train_images_array.shape)
print("Shape of encoded training labels:", train_labels_encoded.shape)
print("Shape of preprocessed test images:", test_images_array.shape)
print("Shape of encoded test labels:", test_labels_encoded.shape)
print("Shape of preprocessed validation images:", val_images_array.shape)
print("Shape of encoded validation labels:", val_labels_encoded.shape)
```

## Define and configure the vision transformer model

### Subtask:
Choose a suitable vision transformer architecture and configure it for this specific task (e.g., number of classes).


**Reasoning**:
Import necessary TensorFlow and Keras components and define the Vision Transformer model architecture, including custom layers for patching and patch encoding, and the overall model structure.



```python
from tensorflow.keras.layers import Input, Conv2D, Dense, Flatten, Reshape, LayerNormalization, Dropout, MultiHeadAttention
from tensorflow.keras.models import Model
from tensorflow.keras import layers

# Define Vision Transformer parameters
input_shape = (128, 128, 3)
patch_size = 16
num_patches = (input_shape[0] // patch_size) * (input_shape[1] // patch_size)
projection_dim = 64
num_attention_heads = 8
transformer_layers = 8
mlp_units = [projection_dim * 2, projection_dim]
num_classes = len(label_encoder.classes_)

# Implement Patches layer
class Patches(layers.Layer):
    def __init__(self, patch_size):
        super().__init__()
        self.patch_size = patch_size

    def call(self, images):
        batch_size = tf.shape(images)[0]
        patches = tf.image.extract_patches(
            images=images,
            sizes=[1, self.patch_size, self.patch_size, 1],
            strides=[1, self.patch_size, self.patch_size, 1],
            rates=[1, 1, 1, 1],
            padding="VALID",
        )
        patch_dims = patches.shape[-1]
        patches = tf.reshape(patches, [batch_size, -1, patch_dims])
        return patches

# Implement PatchEncoder layer
class PatchEncoder(layers.Layer):
    def __init__(self, num_patches, projection_dim):
        super().__init__()
        self.num_patches = num_patches
        self.projection = layers.Dense(units=projection_dim)
        self.position_embedding = layers.Embedding(
            input_dim=num_patches, output_dim=projection_dim
        )

    def call(self, patch):
        positions = tf.range(start=0, limit=self.num_patches, delta=1)
        encoded = self.projection(patch) + self.position_embedding(positions)
        return encoded

# Build the Vision Transformer model
def create_vit_classifier():
    inputs = Input(shape=input_shape)
    # Create patches
    patches = Patches(patch_size)(inputs)
    # Encode patches
    encoded_patches = PatchEncoder(num_patches, projection_dim)(patches)

    # Create Transformer blocks
    for _ in range(transformer_layers):
        # Layer normalization 1
        x1 = LayerNormalization(epsilon=1e-6)(encoded_patches)
        # Create a multi-head attention layer
        attention_output = MultiHeadAttention(
            num_heads=num_attention_heads, key_dim=projection_dim, dropout=0.1
        )(x1, x1)
        # Skip connection 1
        x2 = layers.Add()([attention_output, encoded_patches])
        # Layer normalization 2
        x3 = LayerNormalization(epsilon=1e-6)(x2)
        # MLP
        x3 = Dense(units=mlp_units[0], activation=tf.nn.gelu)(x3)
        x3 = Dropout(0.1)(x3)
        x3 = Dense(units=mlp_units[1], activation=tf.nn.gelu)(x3)
        x3 = Dropout(0.1)(x3)
        # Skip connection 2
        encoded_patches = layers.Add()([x3, x2])

    # Create a [batch_size, projection_dim] tensor
    representation = LayerNormalization(epsilon=1e-6)(encoded_patches)
    representation = Flatten()(representation)
    representation = Dropout(0.5)(representation)
    # Add MLP layers for classification
    features = Dense(units=2048, activation=tf.nn.gelu)(representation)
    features = Dropout(0.5)(features)
    # Classify outputs
    logits = Dense(units=num_classes, activation="softmax")(features)
    # Create the Keras model
    model = Model(inputs=inputs, outputs=logits)
    return model

# Create an instance of the Vision Transformer model
vit_model = create_vit_classifier()

vit_model.summary()
```

## Compile and train the model

### Subtask:
Compile the model with an appropriate optimizer, loss function, and metrics, and then train it using the training data.


**Reasoning**:
Compile the Vision Transformer model with the specified optimizer, loss function, and metrics, and then train it using the preprocessed training data, validating with the validation data.



```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
loss = tf.keras.losses.SparseCategoricalCrossentropy()
metrics = ['accuracy']

vit_model.compile(optimizer=optimizer, loss=loss, metrics=metrics)

history = vit_model.fit(
    train_images_array,
    train_labels_encoded,
    epochs=10,  # Using a reasonable number of epochs
    validation_data=(val_images_array, val_labels_encoded)
)
```

## Evaluate the model

### Subtask:
Evaluate the trained model's performance on the test and validation datasets.


**Reasoning**:
Evaluate the trained model on the test and validation datasets and print the results.



```python
test_loss, test_accuracy = vit_model.evaluate(test_images_array, test_labels_encoded, verbose=0)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")

val_loss, val_accuracy = vit_model.evaluate(val_images_array, val_labels_encoded, verbose=0)
print(f"Validation Loss: {val_loss:.4f}")
print(f"Validation Accuracy: {val_accuracy:.4f}")
```
